```python
import requests
```

```python
page = 1
results = []
while True:
    resp_page = requests.get("https://stepik.org:443/api/courses?page={}".format(page))
    if resp_page.status_code == 200:
        result = [{'id': r['id'],
                   'title': r['title'],
                   'summary': r['summary'],
                   'language': r['language'],
                   'instructors': r['instructors']} for r in resp_page.json()['courses']]
        results.extend(result)
        page += 1
    else:
        break
```

```python
import codecs, json
with codecs.open('data.json', 'w', 'utf8') as f:
     f.write(json.dumps(results, sort_keys = True, ensure_ascii=False))
```
